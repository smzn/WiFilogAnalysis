```python
import numpy as np
import pandas as pd
```

```python
#APファイルの取り込み
df_ap = pd.read_csv('./wifidata/after/APlocations_group.csv')
group_number = df_ap.iloc[-1]['buildID']
print(group_number)
group_transition = np.zeros((group_number, group_number))
df_ap
```

```python
#transition_201XXX_kmeans_X.csvのAP->Groupへの変換
month = '201409'
for cluster in range(5):
    print('cluster = {0}'.format(cluster))
    df_transition = pd.read_csv('./wifidata/cluster/'+month+'/transition_'+month+'_kmeans_'+str(cluster)+'.csv',index_col=0)
    #グループに変換
    transition = df_transition.to_numpy()
    for i in range(len(transition)):
        print('i = {0}, from_group = {1}'.format(i, df_ap.iloc[i]['buildID']))
        for j in range(len(transition)):
            group_transition[df_ap.iloc[i]['buildID']-1][df_ap.iloc[j]['buildID']-1] += transition[i][j]

    pd.DataFrame(group_transition).to_csv('./wifidata/cluster/'+month+'/group_transition_'+month+'_kmeans_'+str(cluster)+'.csv')
    #行和で割る
    #group_transition_rate = group_transition
    #tsum = group_transition.sum(axis=1)
    #for c in group_transition.columns:
    #for c in range(len(group_transition)):
    #    group_transition_rate[c] = group_transition_rate[c]/tsum
        #group_transition_rate = group_transition_rate.fillna(0)
    #pd.DataFrame(group_transition_rate).to_csv('./wifidata/cluster/'+month+'/group_transition_rate_'+month+'_kmeans_'+str(cluster)+'.csv')
```

```python
#duration_201XXX_kmeans_X.csvのAP->Groupへの変換
month = '201401'
for cluster in range(5):
    print('cluster = {0}'.format(cluster))
    df_duration = pd.read_csv('./wifidata/cluster/'+month+'/duration_'+month+'_kmeans_'+str(cluster)+'.csv',index_col=0)
    #グループに変換
    for i, row in enumerate(df_ap.itertuples()):
        df_duration.loc[df_duration['from'] == i, 'from'] = row.buildID
        df_duration.loc[df_duration['to'] == i, 'to'] = row.buildID
        #print(str(i+1)+'/'+str(len(df_ap)))
    df_duration.to_csv('./wifidata/cluster/'+month+'/group_duration_'+month+'_kmeans_'+str(cluster)+'.csv')
```

```python
#クラスタごとに平均系内人数を求める
def getGroupAggregation(group_duration, month, cluster):
    clients_list = []
    node_list = []
    duration_list = []
    duration_sum = 0
    node = 0
    for row in group_duration.itertuples():
        if pd.isnull(row[2]): #取り出した行がNULLの場合はcontinue
            continue
        elif int(row[2]) == 49:#外部から到着
            #print('outside')
            client = row[1] #client
            node = int(row[3]) #toを入れておく
            duration_sum = 0 #0に初期化
        elif int(row[3]) == node: #同じノードへの推移
            duration_sum += float(row[4])
        elif int(row[3]) != node: #他のノードへ移動
            duration_sum += float(row[4]) #たしてからリストに入れる
            clients_list.append(row[1])#リストへ登録
            node_list.append(node)
            duration_list.append(duration_sum)
            node = int(row[3]) #ノードの更新
            duration_sum = 0
         
    group_duration_aggregation = pd.DataFrame({'client': clients_list, 'node': node_list, 'duration': duration_list})
    group_duration_aggregation.to_csv('./wifidata/cluster/'+month+'/group_duration_aggregation_'+month+'_kmeans_'+str(cluster)+'.csv')
    
    #集約したデータフレームから拠点毎の平均滞在時間を算出
    #groupのIDは1~48,49は外部、0は使っていない
    node_number = 50
    limit = 3600 * 3 #(滞在時間は最大3時間とする)
    staytime = np.zeros(node_number)
    staycount = np.zeros(node_number)
    meanstay = np.zeros(node_number)
    for row in group_duration_aggregation.itertuples():
        if row.duration <= limit:
            staytime[int(row.node)] += float(row.duration)
            staycount[int(row.node)] += 1
    for i, val in enumerate(staycount):
        if val > 0:
            meanstay[i] = staytime[i] / val
    pd.DataFrame(meanstay).to_csv('./wifidata/cluster/'+month+'/group_mean_duration_'+month+'_kmeans_'+str(cluster)+'.csv')      
```

```python
month = '201401'
for cluster in range(5):
    df_group_duration = pd.read_csv('./wifidata/cluster/'+month+'/group_duration_'+month+'_kmeans_'+str(cluster)+'.csv')
    getGroupAggregation(df_group_duration, month, cluster)
```

```python
#推移確率 (対角要素を除いたものを作成)
month = '201409'
cluster = 4
transition = np.array(pd.read_csv('./wifidata/cluster/'+month+'/group_transition_'+month+'_kmeans_'+str(cluster)+'.csv',index_col=0))
#対角要素を0
for i in range(len(transition)):
    for j in range(len(transition)):
        if i == j:
            transition[i][j] = 0
#行和を計算
row_sum = transition.sum(axis=1)
list0 = [] #行和が0になるリスト
for i, val in enumerate(row_sum):
    if val == 0:
        list0.append(i)
transition = np.delete(transition, list0, 0) #行を削除
transition = np.delete(transition, list0, 1) #列を削除
print(transition)
row_sum = transition.sum(axis=1)#もう一度行和
print(row_sum)
for i in range(len(transition)):
    transition[i] = transition[i] / row_sum[i]
pd.DataFrame(transition).to_csv('./wifidata/cluster/'+month+'/group_transition_rate_'+month+'_kmeans_'+str(cluster)+'.csv')   
```
