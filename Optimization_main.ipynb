\begin{eqnarray} 
  minimize && \sigma(a) = \frac{1}{|S|}\sum_{i \in S}\sum_{j \in C}(\pi_i^{(j)}(a^{(j)}) - \bar{\pi})^2 \\ 
  subject \ to && a_i^{(j)} \geq 0, a_i^{(j)} \in \mathbb{R}, i \in S, j \in C \nonumber \\
  && \sum_{i \in S}\sum_{j \in C}\frac{1}{a_i^{(j)}} = K^{(j)}, K^{(j)} \in \mathbb{R} \nonumber \\
  && a^{'(j)}(1-\gamma) \leq a^{(j)} \leq a^{'(j)} (1 + \gamma), \gamma \in \mathbb{R} \nonumber
\end{eqnarray}
