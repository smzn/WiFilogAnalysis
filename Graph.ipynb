```python
from graphviz import Digraph
import pandas as pd
```

```python
class GraphTheory_lib :

    def __init__(self, name, shape, color, line, staytime):
        self._G = Digraph(format='png')
        self._G.attr('node', shape='circle')
        self._name = name
        self._shape = shape
        self._color = color
        self._line = line
        self._staytime = staytime

    def getGraph(self, df):
#        N = 16
#        for i, j in zip(range(N), self._name):
#            self._G.node(str(i), str(j))
            #self._G.node(str(i), str(i))
        #self._G.attr('node', shape='square')
        for i, j, k, s in zip(self._name, self._shape, self._color, self._staytime):
            #self._G.node(self._name[0], shape = self._shape[0])
            #滞在時間での分類
            print(s)
            if s < 600 :
                s_idx = 0
            elif s < 1200 :
                s_idx = 1
            elif s < 1800 :
                s_idx = 2
            else :
                s_idx = 3
            print(s_idx)

            #self._G.node(i, shape = j, fillcolor= k, style="filled, solid")
            self._G.node(i, shape = j, fillcolor= k, style= self._line[s_idx])
            #self._G.node(i, shape = j, fillcolor= k, style="dashed", style="filled")

        column = 0
        for column_name, item in df.iteritems(): #1列ずつ取り出す
            if(column > 0): #1列目はインデックスなのでスルー
                print('column_name')
                #print(column_name)
                #print(self._name[column -1])
                print(self._name[int(column_name)])
                print('item')
                for row, value in enumerate(item):
                    print(self._name[int(row)])
                    print(value)
                    if(value > 0.01): #絶対値の場合(100は調整)
                        #self._G.edge(self._name[int(row)], self._name[int(column_name)], label = str(value), penwidth = str(value/1000)) # 絶対値の場合(1000は調整)
                        self._G.edge(self._name[int(row)], self._name[int(column_name)], label = str(round(value,2)), penwidth = str(value * 3)) # 確率の場合
                        #self._G.edge(tail_name, head_name, label, _attributes)
            column += 1

#        for i in range(N):
#            if (i - 1) // 2 >= 0:
#                self._G.edge(str((i - 1) // 2), str(i))

        print(self._G)
        self._G.render('/content/drive/MyDrive/研究/WiFiLog/graph', view=True)

    def getCSV(self, file):
        return pd.read_csv(file, engine='python')
```
